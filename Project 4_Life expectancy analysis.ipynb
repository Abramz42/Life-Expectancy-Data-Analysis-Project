# Exploratory Analysis and Visualization

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=data.loc[0][2:])
plt.xticks(rotation=75)
plt.ylabel('Age')
plt.xlabel('year')
plt.grid()
plt.title('Aruba life expctancy age over the years');
```

```python
plt.figure(figsize=(14,7))
sns.countplot(data=data.loc[0][2:], x=data.loc[0][2:])
plt.title('Age distribution')
plt.xlabel('Age');
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=data.loc[0][2:], label=data['Country Name'][0])
sns.lineplot(data=data.loc[1][2:], label=data['Country Name'][1])
sns.lineplot(data=data.loc[2][2:], label=data['Country Name'][2])
plt.xticks(rotation=75)
plt.title('Aruba, Africa eastern and southern and Afghanistan life expectancy over the years')
plt.ylabel('Age')
plt.xlabel('year')
plt.grid();
```

```python
a = data.loc[0:2][2:].mean(numeric_only=True).astype('int')
plt.figure(figsize=(14,7))
sns.lineplot(data=a)
plt.grid()
plt.ylabel('Age')
plt.xlabel('year')
plt.title('Mean life expectancy age over the years for Aruba, Africa eastern and southern and Afghanistan')
plt.xticks(rotation=75);
```

```python
plt.figure(figsize=(14,7))
sns.countplot(data=a, x=a)
plt.title('Mean age distribution for Aruba, Africa eastern and southern and Afghanistan')
plt.xlabel('Age');
```

```python
africa = data[data['Country Name'] == 'Nigeria'].index
asia = data[data['Country Name'] == 'China'].index
europe = data[data['Country Name'] == 'Russian Federation'].index
north_america = data[data['Country Name'] == 'United States'].index
south_america = data[data['Country Name'] == 'Brazil'].index
oceania = data[data['Country Name'] == 'Australia'].index
continents = list((africa, asia, europe, north_america, south_america, oceania))
continents
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=data.loc[africa[0]][2:], label=data['Country Name'][africa[0]])
sns.lineplot(data=data.loc[asia[0]][2:], label=data['Country Name'][asia[0]])
sns.lineplot(data=data.loc[europe[0]][2:], label=data['Country Name'][europe[0]])
sns.lineplot(data=data.loc[north_america[0]][2:], label=data['Country Name'][north_america[0]])
sns.lineplot(data=data.loc[south_america[0]][2:], label=data['Country Name'][south_america[0]])
sns.lineplot(data=data.loc[oceania[0]][2:], label=data['Country Name'][oceania[0]])
plt.xticks(rotation=75)
plt.legend(title='Country', bbox_to_anchor =(1, 1))
plt.ylabel('Age')
plt.xlabel('year')
plt.title('Life expectancy for the most populous country in each continent over the years')
plt.grid();
```

```python
b=pd.DataFrame()
for num in continents:
    c=data.loc[num]
    b=pd.concat([b,c])
    
b
```

```python
d=b.mean(numeric_only=True).astype('int')
d
```

```python
plt.figure(figsize=(14,7))
sns.lineplot(data=d)
plt.xticks(rotation=75)
plt.ylabel('Age')
plt.title('Mean life expectancy age of all the populous country together in each continent over the years')
plt.xlabel('Year')
plt.grid();
```

```python
plt.figure(figsize=(14,7))
sns.countplot(data=d, x=d)
plt.title('Mean age distribution for the most populous country in each continent')
plt.xlabel('Age');
```

# Questions and Answers

## 1. In the 2020, which country had the highest life expectancy?

```python
e=data[data['2020'] == (data['2020'].max())]
e
```

```python
print('''
In the 2020, which country had the highest life expectancy?
Answer: {}'''.format(e['Country Name'].value_counts().index))
```

## 2. What country had the least life expectancy in the year 2015?

```python
f = data[data['2015'] == (data['2015'].min())]
f
```

```python
print("""
What country had the least life expectancy in the year 2015?
Answer: {}""".format(f['Country Name'].value_counts().index))
```

## 3. What country had the highest increase in life expectancy from 1960 to 2020?

```python
data['Life_Expectancy_growth'] = data['2020'] - data['1960']
data.head(2)
```

```python
g = data[data['Life_Expectancy_growth'] == data['Life_Expectancy_growth'].max()]
g
```

```python
print("""
What country had the highest increase in life expectancy from 1960 to 2020?
Answer: {}""".format(g['Country Name'].unique()[0]))
```

## 4. What's the average life expectancy age generally for the year 2020?

```python
h = data['2020'].mean().astype('int')
h
```

```python
print("""
What's the average life expectancy age generally for the year 2020?
Answer: {}""".format(h))
```
